```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from plot_utils import set_plot_style, set_spines_black
```

```python
# Load the data (.dat FILE))
x = np.loadtxt("DAT_Files/Example_2.dat")
df = pd.DataFrame(x)
print(df.head(5))
```

```python
print(df.nunique())
```

```python
# Identify the indices where the 4th column equals 1
ss1 = np.where(x[:, 3] == 1)[0]
# print("All row indices for stable branch:\n",ss1)
```

```python
print('Index where stable branch starts:', ss1[1])
print('Index where stable branch ends:', ss1[-1])
```

```python
param = x[ss1[1]:ss1[-1] + 1]
param_positive = param[param[:, 0]>0]
```

```python
# Initialize Figure
set_plot_style(dpi = 100)
plt.figure(figsize=(7, 5))


plt.plot(
    param_positive[:, 0],  
    param_positive[:, 1],  
    linestyle='-',
    color='r',
    linewidth=3,
    label='SS'
)


# Grid, labels, and styling
plt.grid(True, which='both', ls='--', lw=0.5)
plt.title('Codimension-one Bifurcation Diagram')
plt.xlabel(r'$\alpha_1$')
plt.ylabel(r'$CDK1^*$')

# Thicken and color the axis spines
ax = plt.gca()
set_spines_black(ax) 
# Add legend
plt.legend(fontsize=13, loc='best', frameon=False)
# Save and show
plt.savefig("Plots_2/CDK1_APC_Bifurcation.pdf", bbox_inches="tight")
plt.show()
```
